***Encoding Categorical Values***

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
dataset['ocean_proximity']= label_encoder.fit_transform(dataset['ocean_proximity'])
```

```python
dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = dataset[['longitude','latitude','housing_median_age','total_rooms',
             'total_bedrooms','population','households',
             'median_income','ocean_proximity']]
y = dataset['median_house_value']

```

```python
X.shape
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression().fit(X_train,y_train)
```

```python
model.score(X_train,y_train),model.score(X_test,y_test),
```

```python
from sklearn.ensemble import RandomForestRegressor
RFModel = RandomForestRegressor(max_depth=20, random_state=0)
RFModel.fit(X_train, y_train)
```

```python
RFModel.score(X_train,y_train),RFModel.score(X_test,y_test)
```
